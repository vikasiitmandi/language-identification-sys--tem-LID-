**Function for accuracy**

```python
def compute_accuracy(y_true, y_pred):
    correct_predictions = 0
    # iterate over each label and check
    for true, predicted in zip(y_true, y_pred):
        if true == predicted:
            correct_predictions += 1
    # compute the accuracy
    accuracy = correct_predictions/len(y_true)
    return accuracy
```

**Accuracy for YouTube data**

```python
YT_true =YT_gt
YT_pred = gmm.predict(df_YT_test_pr2)
# get the accuracy
compute_accuracy(YT_true, YT_pred)
```

**Accuracy for Prasharbharti data**

```python
PB_true =PB_gt
PB_pred = gmm.predict(df_PB_test_pr2)
# get the accuracy
compute_accuracy(PB_true, PB_pred)
```

**AIC values**
